```python
import math
```

```python
H_0 = 1.0 #  --1.3  0.7
h_0 = 5 #
g = 9.7863 #  
c_0 = math.sqrt(g*(h_0+H_0)) #

distance_10 = 3000 #3.0km --- 10m
distance_5 = 1000 # 800m --- 5m
ocean_k = h_0/distance_5 

beach_width = 60 #m
beach_height = 2.4 #m
#beach_angle = 5*math.pi/180
beach_k = beach_height/beach_width
```

```python
def sech(x):
    return 1/math.cosh(x)

def Ocean_Wave(x,t,H,h):
    h3 = math.pow(h,3)
    a1 = math.sqrt(0.75*H/h3)
    c = math.sqrt(g*(h+H)) #
    X = a1*(x-c*t)
    #print(X)
    return H*math.pow(sech(X),2)    

def Wave_V(x,z,t,H,h):
    h3 = math.pow(h,3)
    a1 = math.sqrt(0.75*H/h3)
    c = math.sqrt(g*(h+H)) #
    a2 = math.sqrt(g*h)*H/h
    a3 = math.sqrt(3*g*h)*math.pow(H/h,1.5)/h
    X = a1*(x-c*t)
    u = a2*math.pow(sech(X),2)
    w = a3*z*math.pow(sech(X),2)*math.tanh(X)
    return u,w
```

```python
h_x = [h_0 - i*ocean_k for i in range(distance_5)]
H_x = [1.0]

def H_near(h_i,H_j):
    n_i0= 0.55+0.3*(h_0-h_i)/ocean_k/distance_5
    c_i = math.sqrt(g*(h_i+H_j))
    k_s = h_0/h_i*math.pow(c_0/c_i*n_i0,3)
    return k_s*H_j

for i in range(len(h_x)):
    if i > 0:
        a=H_near(h_x[i],H_x[i-1])
        H_x.append(a)

plt.plot(H_x)
plt.title('the H(waveform height) of the distance of the 5m-line ')
plt.ylabel('H(waveform height) (m)')
plt.xlabel('the distance of the 5m-line (m)')
#plt.xlim(0,860)
#plt.ylim(0,20)
plt.savefig('the H(waveform height) .png')
```

```python
def wave(pos_x=1):
    h3 = math.pow(h_x[pos_x],3)
    a1 = math.sqrt(0.75*H_x[pos_x]/h3)
    c = math.sqrt(g*(h_x[pos_x]+H_x[pos_x]))
    #print('c:',math.sqrt(g*(h_x[pos_x]+H_x[pos_x])))
    #print('H:',H_x[pos_x])
    #print('a:',a1)
    t_max = pos_x/c
    spectrum = 300/(a1*c)/2
    t = np.linspace(t_max-spectrum,t_max+spectrum,200)
    y=[]
    for i in range(len(t)):
        y.append(Ocean_Wave(pos_x,t[i],H_x[pos_x],h_x[pos_x]))
    plt.plot(t,y)
    plt.title('the wave function of 800m ')
    plt.xlabel('t(s)')
    plt.ylabel('wave function(m)')
    plt.savefig('the wave function of 800m.png')
```

```python
def wave_uw(pos_x=1):
    h3 = math.pow(h_x[pos_x],3)
    a1 = math.sqrt(0.75*H_x[pos_x]/h3)
    c = math.sqrt(g*(h_x[pos_x]+H_x[pos_x]))

    t_max = pos_x/c
    spectrum = 300/(a1*c)/2
    t = np.linspace(t_max-spectrum,t_max+spectrum,200)
    
    u=[]
    w=[]
    for i in range(len(t)):
        z = Ocean_Wave(pos_x,t[i],H_x[pos_x],h_x[pos_x])
        u0,w0=Wave_V(pos_x,z,t[i],H_x[pos_x],h_x[pos_x])
        u.append(u0)
        w.append(w0)
    plt.plot(t,u)
    #plt.plot(t,w)
    plt.legend(['level_v','vertical_v'])
```

```python

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#
data = [187,204,213,209,194,174,142,115,89,66,52,52,59,67,80,98,100,105,109,113,117,125,143,158,176,196]
data = [i/100 for i in data]

def f_1(x):
    a0 = 124.8
    a1 = 64.05
    b1 = 14.88 
    a2 = -0.5196 
    b2 =  27.36 
    w  =  0.2559  
    result = a0 + a1*math.cos(x*w) + b1*math.sin(x*w) + a2*math.cos(2*x*w) + b2*math.sin(2*x*w)
    return result/100

def f_2(x):
    a0 =  137.1  
    a1 =  70.66 
    b1 =  -18.31 
    w  =  0.2156
    result =  a0 + a1*math.cos(x*w) + b1*math.sin(x*w)
    return result/100


x = np.linspace(0,25,250)
res1=[]
res2=[]
for i in x:
    res1.append(f_1(i))
    res2.append(f_2(i))
plt.plot(data,'*')
plt.plot(x,res1,linewidth = 1.0)
plt.plot(x,res2,linewidth = 1.0)
plt.legend(['raw data','function_1','function_2'])
plt.xlabel('time')
plt.ylabel('height of tide')
plt.title('the fitted cureve of height of tide ')
plt.savefig('the fitted cureve of height of tide.png')
plt.show()
```

```python
sand_r = 0.002 #2mm
sand_p = 2.65*1000 #
sand_n0 = 10 #
snad_v0 = 0.2 #
sand_m0 = 0.4 #
water_v = 6 #

cell_d = sand_r*2*5 #5x5x5
cell_mass = math.pow(cell_d,3)*sand_p #
```

```python
#
def p_fun(x):
    #
    p_u=0.3
    p_a=0.6
    a=1/(math.sqrt(2*np.pi)*p_a*x)
    c=-0.5*math.pow(math.log(x)-p_u,2)/math.pow(p_a,2)
    b=np.exp(c)
    res = a*b
    return res
#
def F_fun(x,n):
    sand_m = sand_m0 * np.exp(-x)
    sand_n = sand_n0 * p_fun(x)
    
    F_f=sand_m*cell_mass*g*n
    F_a=np.pi*6*sand_n*sand_r
    F_s=snad_v0/(1+snad_v0)/2*cell_mass*g*n
    F = F_f + F_a - F_s
    return F,F_f,F_a

#
#
def P_water(x):
    if x == 0:
        return 0.57
    else:
        a,b,c=F_fun(x,1)
        return (0.35-a)*2
```

```python
from matplotlib.pyplot import MultipleLocator
F_0 = sand_m0*cell_mass*g + np.pi*6*sand_n0*sand_r - snad_v0/(1+snad_v0)/2*cell_mass*g
x = np.linspace(0,100,1000)
y = [F_0]
y2 = [sand_m0*cell_mass*g ]
y3 = [np.pi*6*sand_n0*sand_r]
for i in x:
    if i > 0 :
        a,b,c=F_fun(i,1)
        y.append(a)
        y2.append(b)
        y3.append(c)
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(['F_total','F_friction','F_viscosity'])
plt.xlim(0.1,10)
x_major_locator=MultipleLocator(1)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
plt.xlabel('water content (%)')
plt.ylabel('resisitibility')
plt.title('resisitibility of different water content')
plt.savefig('resisitibility of different water content.png')
```
